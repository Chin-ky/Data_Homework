```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
# selected_features = df[['names', 'of', 'selected', 'features', 'here']]
selected_features = df.drop(columns = "koi_disposition")
selected_features.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df.koi_disposition
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(selected_features, y, random_state=1)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a random forest classifier
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier(n_estimators=200)
model2 = model2.fit(X_train_scaled, y_train)
model2.score(X_test_scaled, y_test)
```

```python
importances = model2.feature_importances_
importances
```

```python
sorted(zip(model2.feature_importances_, selected_features), reverse=True)
```

# Train the Model



```python
print(f"Training Data Score: {model2.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model2.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.svm import SVC 
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model
```

```python
# Create the GridSearch estimator along with a parameter object containing the values to adjust
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid2 = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid2.fit(X_train_scaled, y_train)
```

```python
print(grid2.best_params_)
print(grid2.best_score_)
```
